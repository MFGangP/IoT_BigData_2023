### 판다스 데이터 합치기

- 데이터 합치는 방법
    - DB JOIN, UNION
        - JOIN - 데이터를 옆으로(가로) 합침 == > pandas.merge
        - UNION - 데이터를 밑으로(세로) 합침 ==> pandas.append(X) -> pandas.concat

```python
import pandas as pd
import numpy as np
```

```python
# 첫번째 데이터 로드
sample1 = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_1.xlsx',
sheet_name='Sheet1', header=1, skipfooter=2)

sample1
```

```python
sample1['입국객수'].sum()
```

```python
sample1['기준년월']='2019-11'
sample1
```

```python
# 코드마스터(DB의 코드 테이블 동일)
code_master = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_codemaster.xlsx')

code_master
```

```python
## DB JOIN = Pandas.merge
result1 = pd.merge(left=sample1, right=code_master, how='left', left_on = '국적코드', right_on='국적코드')
result1
```

```python
## DB JOIN = Pandas.merge
result2 = pd.merge(left=sample1, right=code_master, how='inner', left_on = '국적코드', right_on='국적코드')
result2
```

```python
## DB UNION = Pandas.concat
sample2 = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_2.xlsx',
                        header=1, skipfooter=2)

sample2['기준년월'] = '2019-12'
sample2
```

```python
total_sample = pd.concat([sample1, sample2])

total_sample
```
